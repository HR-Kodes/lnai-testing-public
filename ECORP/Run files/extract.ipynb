```python
structured_data=[]
for address in df2:
    # Set the input data
    input_data = {
        "source": "text",
        "value": address,
        "content_type": "application/json"
    }
    response = requests.post(
        endpoint + "/formrecognizer/v2.1-preview.3/prebuilt/address/analyze",
        headers={
            "Ocp-Apim-Subscription-Key": subscription_key,
            "Content-Type": input_data["content_type"]
        },
        json=input_data["value"]
    )
    structured_data.append(response.json()["analyzeResult"]["documentResults"][0]["fields"])
```
