```python
import pandas as pd
import ast
# Read the Excel file
# file_path = r"C:\Users\golup\Downloads\Mumbai2.xlsx"
file_path=r'Mumbai3.xlsx'
df = pd.read_excel(file_path)
  
```

```python
df['C_NAME1'][83]
```

```python
f="['All', 'All'"
g="['All','All'"

def convert_array(row):
    data = row['C_NAME1']
    if g in data:
        # data = data.replace(",","],[")
        row['C_NAME1'] = data
        data = data.replace(",","],[")
        
        return row
    if f in data:
        # data = data.replace(", ",",],[")
        data = data.replace(",","],[")
        data = data.replace(" '","'")
        
        
        row['C_NAME1'] = data
        return row
    else:
        return row
    
df1=df.apply(lambda row : convert_array(row),axis=1)
```

```python
df1.to_excel("Mumbai3.xlsx")
```
