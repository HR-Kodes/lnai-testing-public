```python
import glob
import os
import cv2
from PIL import Image
os.chdir(r'C:\Users\mrock\Downloads\test foldder\extraction\plates')
# os.
# glob.glob()
detect_list = []
lables = glob.glob('labels\*.txt')
vid = glob.glob('*.mp4')[0]
# print(vid)
for i in lables:
    frame = int(i.split('.')[0].split('_')[-1])
    with open(i,'r') as f:
        labels = f.readlines()
    video = cv2.VideoCapture(vid)
    # print(type(frame))
    video.set(cv2.CAP_PROP_POS_FRAMES,frame)
    r, img= video.read()
    # print(r)
    if r:
        color_coverted = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(color_coverted)
        height  = img.height
        width = img.width
        # detect_list = []
        for i in labels:
            value = [float(a) for a in i.split()]
            sw = int(width * float(value[1] - value[3]/2))
            ew = int(width * float(value[1] + value[3]/2))
            sh = int(height * float(value[2] - value[4]/2))
            eh = int(height * float(value[2] + value[4]/2))
            detect_list.append([img.crop((sw, sh, ew, eh)),value[5]])
# img.show()
```

```python
import matplotlib.pyplot as plt
for i,j in enumerate(detect_list):
    plt.subplot(len(detect_list),1,i+1)
    plt.imshow(j[0])
    plt.title(j[1])

plt.show()
```
