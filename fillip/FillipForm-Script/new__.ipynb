```python
def upload_file_from_url(session_token, file_url, ref_no, filename, folder, attachment_category, attachment_label):
    response = requests.get(file_url)
    if response.status_code != 200:
        print(f'Failed to download file from URL: {file_url}')
        return
    headers = {
        'Cookie': f'cookiesession1=678B286955776DB15177E7C622CE6403; alertPopup=true; session-token={session_token}',
        'Origin': 'https://www.mca.gov.in',
    }
    payload = {
        'file': (filename, response.content, 'image/jpeg'),
        'folder': folder,
        'filemetadata': f'{{"form":"FiLLiP","srn":"{ref_no}","documentType":"Attachment","module":"FiLLiP","userid":"FiLLiP","applicationId":"{ref_no}","attachmentCategory":"{attachment_category}","attachmentLabel":"{attachment_label}"}}'
    }
    url = 'https://www.mca.gov.in/bin/mca/dms/addservice'
    upload_response = requests.post(url, headers=headers, files=payload)
    if upload_response.status_code == 200:
        print('File uploaded successfully')
        print('Upload Response: ', upload_response.text)
        return upload_response.text
    else:
        print('Failed to upload the file')
```

```python
import time
import json

def upload_file_with_retry(session_token, s3_url, ref_no, filename, folder, attachment_category, attachment_label, max_retries=3):
    for attempt in range(max_retries):
        try:
            response = json.loads(upload_file_from_url(session_token, s3_url, ref_no, filename, folder, attachment_category, attachment_label))
            response['folderName'] = response['folderId']
            return response
        except Exception as e:
            if attempt < max_retries - 1:
                print(f"Upload attempt {attempt + 1} failed. Retrying in 5 seconds...")
                time.sleep(5)
            else:
                print(f"Upload failed after {max_retries} attempts.")
                raise e  # Raise the exception if max retries reached

# Example usage for each upload:

# Address Proof Attachment
s3_url_1 = inp['url']['3eAdressProofURL']
filename_1 = s3_url_1.split('/')[-1]
attachment_category_1 = s3_url_1.split('.')[-1]
att_resp1 = upload_file_with_retry(session_token, s3_url_1, ref_no, filename_1, "", attachment_category_1, "3eAdressProof")

# Utility Bills Attachment
s3_url_2 = inp['url']['3eUtilityBillsURL']
filename_2 = s3_url_2.split('/')[-1]
attachment_category_2 = s3_url_2.split('.')[-1]
att_resp2 = upload_file_with_retry(session_token, s3_url_2, ref_no, filename_2, "", attachment_category_2, "3eUtilityBills")

# Identity Proof Attachment 1
s3_url_3 = inp['url']['6bIdentityProof1URL']
filename_3 = s3_url_3.split('/')[-1]
attachment_category_3 = s3_url_3.split('.')[-1]
att_resp3 = upload_file_with_retry(session_token, s3_url_3, ref_no, filename_3,"", attachment_category_3, "6bIdentityProof")

# Residential Proof Attachment 1
s3_url_4 = inp['url']['6bResidentialProof1URL']
filename_4 = s3_url_4.split('/')[-1]
attachment_category_4 = s3_url_4.split('.')[-1]
att_resp4 = upload_file_with_retry(session_token, s3_url_4, ref_no, filename_4, "", attachment_category_4, "6bResidentialProof")

# Identity Proof Attachment 2
s3_url_5 = inp['url']['6bIdentityProof2URL']
filename_5 = s3_url_5.split('/')[-1]
attachment_category_5 = s3_url_5.split('.')[-1]
att_resp5 = upload_file_with_retry(session_token, s3_url_5, ref_no, filename_5, "", attachment_category_5, "6bIdentityProof")

# Residential Proof Attachment 2
s3_url_6 = inp['url']['6bResidentialProof2URL']
filename_6 = s3_url_6.split('/')[-1]
attachment_category_6 = s3_url_6.split('.')[-1]
att_resp6 = upload_file_with_retry(session_token, s3_url_6, ref_no, filename_6, "", attachment_category_6, "6bResidentialProof")
```
