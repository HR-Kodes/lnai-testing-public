```python
#for din details
import requests
import json

# Define the URL
url = "https://www.mca.gov.in/bin/MDSMasterDataServlet"

# Define the headers
headers = {
    "Host": "www.mca.gov.in",
    "Cookie": "cookiesession1=678B28693F2E8C95FB25BB5BA8BC843B",
    "Content-Type": "application/json",
}

# Define the JSON payload
payload = {
    "ID": "09730656",
    "requestID": "din"
}

# Send the POST request
response = requests.post(url, headers=headers, data=json.dumps(payload))

# Check the response status code
if response.status_code == 200:
    # Request was successful
    print("Request was successful.")
    print("Response JSON:")
    print(response.json())  # Parse response JSON if available
else:
    # Request failed
    print(f"Request failed with status code {response.status_code}")
    print("Response content:")
    print(response.text)  # Print response content for debugging
```
