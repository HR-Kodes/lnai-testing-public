```python
import glob
import json
from azure.core.credentials import AzureKeyCredential
from azure.ai.formrecognizer import DocumentAnalysisClient

files = glob.glob('*.pdf')


for f in files:
    endpoint = "https://gstocr.cognitiveservices.azure.com/"
    key = "55fc3631297d4d7489b926b6d84ec2e1"

    model_id = "gst-ocr"
    # formUrl = "YOUR_DOCUMENT"

    document_analysis_client = DocumentAnalysisClient(
        endpoint=endpoint, credential=AzureKeyCredential(key)
    )
    print(f"labeling  {f}")
    with open(f, "rb") as file:
        document_analysis_client = DocumentAnalysisClient(
            endpoint=endpoint, credential=AzureKeyCredential(key)
        )
        poller = document_analysis_client.begin_analyze_document(model_id, file)
        result = poller.result()
        result_data={}
    for idx, document in enumerate(result.documents):
        for key, value in document.fields.items():
            result_data[key] = value.value if value.value else value.content
            if key == 'ITEM':
                result_data[key] = {}
                if value.value:
                    for i,j in value.value[0].value.items():
                        result_data[key][i]=j.value

    # Convert the dictionary to JSON format
    json_data = json.dumps(result_data, indent=2)
    with open(f[:-4]+'.json','w') as t:
        t.write(json_data)
```
