```python
import requests
import base64

# First part: Fetching captcha images and saving them
def fetch_captcha_image():
    url = "https://ipindiaonline.gov.in/trademarkefiling/CaptchaGenerator/captcha.aspx?638273722548402461"
    headers = {
        # Your headers here
        "Cookie": "ASP.NET_SessionId=00g41w2wxba1z2ue0ryv4tll"
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        with open("captcha_image.jpg", "wb") as image_file:
            image_file.write(response.content)
        print("Captcha image saved successfully.")
    else:
        print("Failed to fetch captcha image.")

# Second part: Processing captcha images using the OCR API
def process_captcha_image():
    url = "https://app.nanonets.com/api/v2/OCR/FullText"
    image_filename = "captcha_image.jpg"

    files = [
        ('file', ('captcha_image.jpg', open(image_filename, 'rb'), 'application/jpeg'))
    ]
    headers = {}

    response = requests.request("POST", url, headers=headers, files=files, auth=requests.auth.HTTPBasicAuth('3edab03f-18ab-11ee-b7d0-86364e72d134', ''))

    return response.json()

# Main loop to fetch and process captchas
num_captchas = 200
for i in range(num_captchas):
    fetch_captcha_image()
    response_data = process_captcha_image()

    if "results" in response_data and len(response_data["results"]) > 0:
        captcha_text = response_data["results"][0]["page_data"][0]["raw_text"]
        print(f"Captcha text: {captcha_text}")
        
        # Save captcha text to data.txt
        with open("data.txt", "a") as data_file:
            data_file.write(captcha_text + "\n")
    else:
        print("Failed to process captcha image.")

print("Completed processing captchas.")
```
