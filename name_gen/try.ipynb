```python
from combine_check import fetch_session_token, check_llp_name, check_private_limited_name
from name_gen import fetch_distinct_names

def main():
    description = "ai powered financial tech"
    session_token = fetch_session_token()
    available_names=[]
    while True:
        company_type = input("Enter company type (llp/private limited): ").lower()

        if company_type in ["llp", "private limited"]:
            break
        else:
            print("Invalid input. Please enter 'llp' or 'private limited'.")

    company_names = fetch_distinct_names(description)

    modified_names = [f"{name} {company_type}" for name in company_names]

    if company_type == "llp":
        check_function = check_llp_name
    else:
        check_function = check_private_limited_name

    for c_name in modified_names:
        if check_function(c_name, session_token):
            available_names.append(c_name)
            print(f"{c_name} is available as {company_type.capitalize()}.")

if __name__ == "__main__":
    main()
```
