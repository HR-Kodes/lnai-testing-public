```python
import hashlib

def str2rstr_utf8(s):
    return s.encode('utf-8')

def rstr_sha1(s):
    sha1 = hashlib.sha1()
    sha1.update(s)
    return sha1.digest()

def rstr2hex(input_bytes):
    return ''.join(format(byte, '02x') for byte in input_bytes)

def hex_sha1(s):
    return rstr2hex(rstr_sha1(str2rstr_utf8(s)))

def calculate_password_hash(password, seed):
    SHA1New = hex_sha1(password)
    finalpass = SHA1New + seed
    SHA1 = hex_sha1(finalpass)
    return SHA1

  # Replace this with the actual value of seed from your code

hashed_password = calculate_password_hash(password, str(seed))
# hashed_password = calculate_password_hash(password, "427215818")

print(hashed_password)
```
